```python
df=df.dropna()
print(df)
```

```python
#print(df.isnull().sum())
print(df)
```

```python
s=pd.get_dummies(df["Sex"],drop_first=True)
```

```python
p=pd.get_dummies(df["Pclass"],drop_first=True)
```

```python
e=pd.get_dummies(df["Embarked"],drop_first=True)
```

```python
df=pd.concat([df,s,p,e],axis=1)
print(df)
```

```python
print(df)
```

```python
df=df.drop(['Cabin'],axis='columns')
print(df)
```

```python
y=df['Survived']
x=df.drop(['Survived'],axis=1)
print(x,y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logr=LogisticRegression(random_state=10)
```

```python
logr.fit(x_train,y_train.values.ravel())
```

```python
predictions=logr.predict(x_test)
```

```python
from sklearn.metrics import classification_report
```

```python
classification_report(y_test,predictions)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,predictions)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("accuracy is",accuracy_score(y_test,predictions)*100,"%")
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gb=GaussianNB()
```

```python
gb.fit(x_train,y_train)
```

```python
y_pred=gb.predict(x_test)
print(y_pred)
```

```python
a=accuracy_score(y_test,y_pred)
print(a)
```
